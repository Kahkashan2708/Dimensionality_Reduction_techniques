```python
import pandas as pd
from sklearn.datasets import load_digits
```

```python
datasets = load_digits()
datasets.keys()
```

```python
df = pd.DataFrame(datasets.data, columns=datasets.feature_names)
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
X= df
y= datasets.target
```

```python
print(y)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(X)
x_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x_scaled,y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
X.shape
```

```python
from sklearn.decomposition import PCA
pca = PCA(0.90) # 90% variance
x_pca = pca.fit_transform(x_scaled)
x_pca.shape
```

```python
x_pca
```

```python
pca.explained_variance_ratio_
```

```python
pca.n_components_
```

```python
X_train_pca, X_test_pca, y_train_pca, y_test_pca = train_test_split(x_pca,y, test_size=0.2, random_state=42)
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train_pca, y_train_pca)
```

```python
model.score(X_test_pca, y_test_pca)
```

### Decreased number of features from 64 to 31, still accuracy is almost same

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(x_scaled)
X_pca.shape
```

```python
pca.explained_variance_ratio_
```

```python
pca.explained_variance_ratio_.sum()
```

```python
X_train_pca, X_test_pca, y_train_pca, y_test_pca = train_test_split(X_pca,y, test_size=0.2, random_state=42)
```

```python
model = LogisticRegression()
model.fit(X_train_pca,y_train_pca)
```

```python
model.score(X_test_pca, y_test_pca)
```

#### Accuracy Decreased

## Visualization

```python
plt.figure(figsize=(8,6))

for digit in range(10):  
    idx = datasets.target == digit  
    plt.scatter(
        X_pca[idx, 0],
        X_pca[idx, 1],
        label=f"Digit {digit}",
        alpha=0.7
    )

plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.title('PCA Projection of Digits Dataset')
plt.legend(title="Classes", bbox_to_anchor=(1.05, 1), loc='upper left')  # Legend outside
plt.tight_layout()
plt.show()
```
